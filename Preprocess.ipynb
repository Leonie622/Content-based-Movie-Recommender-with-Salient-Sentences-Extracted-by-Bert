```python
import pandas as pd
```

```python
corpus_df = pd.read_csv("./corpus_cleaned.csv")
corpus_df.head()
```

```python
corpus_df
```

```python
def merge_corpus(row):
    movielist = eval(row["corpus_movie"])
    trailerlist = eval(row["corpus_trailer"])
    # merge lines with less than 10 words to he previous line
    new_trailer = []
    for sentence in trailerlist:
        if not new_trailer:
            new_trailer.append(sentence)
            continue
        if len(sentence) < 10:
            new_trailer[len(new_trailer)-2] += sentence
        else:
            new_trailer.append(sentence)
    new_trailer = ['@highlight '+sentence for sentence in new_trailer]
    trailerlist = new_trailer
    return list(trailerlist) + list(movielist)

corpus_df["merged_corpus"] = corpus_df.apply(lambda row:merge_corpus(row), axis=1)
```
